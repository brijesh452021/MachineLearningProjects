```python
df=pd.read_csv("iris.csv")
df.head()
```

```python
df_setosa=df.loc[df["Iris-setosa"]=="Iris-setosa"]
df_setosa.head()
```

```python
df["Iris-setosa"].unique()
```

```python
df_versicolor=df.loc[df["Iris-setosa"]=="Iris-versicolor"]
df_virginica=df.loc[df["Iris-setosa"]=="Iris-virginica"]
```

**Univariate Analysis**


```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(df_setosa["3.5"],np.zeros_like(df_setosa["3.5"]),'o')
plt.plot(df_versicolor["3.5"],np.zeros_like(df_versicolor["3.5"]),'o')
plt.plot(df_virginica["3.5"],np.zeros_like(df_virginica["3.5"]),'o')
```

**Bivariate Analysis**

```python
import seaborn as sns
x=sns.FacetGrid(df,hue="Iris-setosa",size=5)
x=x.map(plt.scatter,"3.5","1.4").add_legend()
```

**Multivariate Analysis**

```python
sns.pairplot(df,hue="Iris-setosa",size=5)
```
